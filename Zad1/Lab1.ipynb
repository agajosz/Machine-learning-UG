```python
import numpy as np
import statistics as st
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from collections import defaultdict

#Urodzenia w Gdańsku w latach 2008-2017 z podziałem na dzielnice
dane = pd.read_csv('urodzenia.csv')

#Dane dla urodzeń w 2008 i wszystkich dzielnic
narodziny = dane[dane["Narodziny czy Zgon"] == "narodziny"]
print("Mean", st.mean(narodziny["Rok 2008"]))
print("Median", st.median(narodziny["Rok 2008"]))
print("Mode", st.mode(narodziny["Rok 2008"]))

devitation = st.pstdev(narodziny["Rok 2008"])
print("Devitation", devitation)

variance = st.pvariance(narodziny["Rok 2008"])
print("Variance", variance)
#===============================================
subset_2008_df = dane[['Dzielnica','Płeć', 'Narodziny czy Zgon', 'Rok 2008']]
#Histogram
plt.hist(subset_2008_df)
```
